```python
import pandas as pd
import os

def xlsx_to_csv(input_file):
    # Check if the input file is an xlsx file
    if not input_file.endswith('.xlsx'):
        print("Input file must be an xlsx file.")
        return
    
    # Load the Excel file
    try:
        df = pd.read_excel(input_file)
    except Exception as e:
        print(f"Error loading Excel file: {e}")
        return
    
    # Determine the output CSV filename
    csv_filename = os.path.splitext(input_file)[0] + '.csv'
    
    # Convert and save to CSV
    try:
        df.to_csv(csv_filename, index=False)
        print(f"CSV file saved: {csv_filename}")
    except Exception as e:
        print(f"Error saving CSV file: {e}")

# Example usage:
xlsx_file = r'Z:\UriMons\Fiber Photometry\adBNST_eEPA_GCaMP8s\1st_batch\Behavior\HeaDipping\M354_Processed\M354_Eating.xlsx'  # Replace with your xlsx file path
xlsx_to_csv(xlsx_file)
```
