```python
from moviepy.editor import VideoFileClip
import os
```

```python
def crop_video(input_video, output_video, start_frame, end_frame):
    # Load the video clip
    video_clip = VideoFileClip(input_video)

    # Calculate the start and end times in seconds
    fps = video_clip.fps
    start_time = start_frame / fps
    end_time = end_frame / fps

    # Crop the video
    cropped_clip = video_clip.subclip(start_time, end_time)

    # Write to MP4 with proper codec settings
    cropped_clip.write_videofile(
        output_video,
        codec='libx264',
        audio_codec='aac',
        temp_audiofile='temp-audio.m4a',
        remove_temp=True
    )

    # Clean up
    video_clip.close()
    cropped_clip.close()

# Example usage
input_folder = r'Z:\UriMons\Fiber Photometry\GCaMP\adBNST_eEPA_GCaMP8s\1st_batch\Experiment\AVI'
input_video_name = 'M329_Cam2.avi'
input_video_path = os.path.join(input_folder, input_video_name)

output_folder = r'E:\Weizmann Institute Dropbox\Uri Monsonego\Lab\Presentations\Lab meeting 2025_11'
output_video_name = 'M329_Cam2_18600-19200f.mp4'  
output_video_path = os.path.join(output_folder, output_video_name)

start_frame = 18600
end_frame = 19200

crop_video(input_video_path, output_video_path, start_frame, end_frame)
```
