Power claculation

```python
from statsmodels.stats.power import TTestIndPower

effect_size = 0.8  # Cohen's d (medium effect)
alpha = 0.05       # Significance level
power = 0.7        # Desired power

analysis = TTestIndPower()
sample_size = analysis.solve_power(effect_size=effect_size, alpha=alpha, power=power, ratio=1)
print(f"Sample size per group: {int(sample_size)}")
```

Cohen's d calculation

```python
import numpy as np

def cohens_d_from_stats(mean_ko, sd_ko, n_ko, mean_wt, sd_wt, n_wt):
    """Calculate Cohen's d given means, standard deviations, and sample sizes for KO and WT groups."""
    pooled_std = np.sqrt(((sd_ko ** 2) * (n_ko - 1) + (sd_wt ** 2) * (n_wt - 1)) / (n_ko + n_wt - 2))
    return (mean_ko - mean_wt) / pooled_std

# Example usage (modify these values for your experiment)
mean_KO = 110  # Mean of KO group
sd_KO = 15     # SD of KO group
n_KO = 10      # Number of mice in KO group

mean_WT = 100  # Mean of WT group
sd_WT = 12     # SD of WT group
n_WT = 10      # Number of mice in WT group

d = cohens_d_from_stats(mean_KO, sd_KO, n_KO, mean_WT, sd_WT, n_WT)
print(f"Cohen's d (aka effect size): {d:.3f}")
```
